## CWSI - 수분 스트레스 지수 


- cwsi
- 작물의 수분에 대한 스트레스 정도를 0 ~ 1 사이로 표현하는 지수
- 작물의 엽온이 대기 온도 보다 1 ~ 4 도 낮은 경우 작물의 층산이 최대가 되고 cwsi가 0
- 작물의 엽온이 대기 온도 보다 1 ~ 4 도 높은 경우 작물의 층산이 최소가 되고 cwsi가 1
